# Movies Data Set Analysis

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)
```

```python
df = pd.read_csv(r'C:\Users\USER\Downloads\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} . {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
# Handle NaN and infinite values
df['budget'] = df['budget'].replace([np.inf, -np.inf], np.nan)  # Replace inf with NaN
df['budget'] = df['budget'].fillna(0)  # Replace NaN with 0 (or another appropriate value)
df['budget'] = df['budget'].astype('Int64') 

# Handle NaN and infinite values
df['gross'] = df['gross'].replace([np.inf, -np.inf], np.nan)  # Replace inf with NaN
df['gross'] = df['gross'].fillna(0)  # Replace NaN with 0 (or another appropriate value)
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
#Look Correlation
df.corr(method='pearson') #pearson, kendall, spaeman
```

```python
#High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Metric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
